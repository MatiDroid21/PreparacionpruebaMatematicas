La latencia puede verse afectada por la velocidad de transmisión de los datos a través de los cables o conexiones inalámbricas, la distancia física entre los dispositivos, el tiempo que tarda un computador en procesar o reenviar los datos, etc.

# Problema 7

El número de usuarios de una red social se puede modelar mediante la función:

$$U(t) = \frac{1.000}{1+9e^{-0.5t}}$$

donde $t$ corresponde al tiempo transcurrido en meses.

1.  Defina variable dependiente e independiente, indicando unidad de medida.
2.  Determine la cantidad de usuarios transcurridos $12$ meses.
3.  Grafique la función, utilizando la librería *Matplotlib*, para los primeros dos años de funcionamiento.
4.  ¿Cuánto tiempo debe pasar para que la red social llegue a $800$ usuarios?

## Solución

**1)**

**Variable dependiente**: Cantidad de usuarios (unidades)

**Variable independiente**: Tiempo transcurrido (meses)

**2)**

```
import numpy as np

def U(t):
  return 1000/(1 + 9 * np.exp(-0.5 * t))

# Transcurridos 12 meses
t = 12
print(f'Transcurridos {t} meses la red social tiene {U(t):.0f} usuarios aproximadamente.')
```

**3)** Gráfico de la función

```
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

fig, ax = plt.subplots()

t = np.arange(0, 24, 0.1)

plt.plot(t, U(t))


plt.title('Relación entre cantidad de usuarios de una red social y el tiempo transcurrido')
plt.xlabel('Tiempo transcurrido (meses)')
plt.ylabel('Cantidad de usuarios (unidades)')


plt.grid(True)
plt.show()
```

**4)**

```
import numpy as np
from scipy.optimize import fsolve

def u(t):
  return 1000 / (1 + 9 * np.exp(-0.5*t)) - 800

t = np.linspace(0, 25, 1)
sol = fsolve(u, t)

print(f'La red social tendrá 800 usuarios transcurridos {sol[0]:.1f} meses aproximadamente.')
```

# Problema 8

$$A(n) = 0,01n^{2} + 0,5n + 2$$

donde $n$ corresponde a la cantidad de elementos a ordenar. El algoritmo funciona desde las 23:00 horas hasta las 06:00 horas y debe ordenar $1.562$ elementos.

1. Defina variable dependiente e independiente, indicando unidad de medida.
2. Determine el dominio contextualizado de la función.
3. Determine el tiempo de ejecución para ordenar $1.200$ elementos.
4. Grafique, utilizando la biblioteca *Matplotlib*, la función $A(n)$.
5. ¿Cuántos elementos ordena luego de $6$ horas de funcionamiento?

## Solución

**1)**

**Variable dependiente**: Tiempo de ejecución (segundos)

**Variable independiente**: Cantidad de elementos (unidades)

**2)**

$Dom$ $A(n): [0;1.562]$

**3)**

```
def A(n):
  return 0.01 * n**2 + 0.5 * n + 2

print(f'El tiempo de ejecución del algoritmo al ordenar 1.200 elementos es {A(1200)} segundos, equivalente a {A(1200)/3600:.1f} horas aproximadamente.')
```

```
import matplotlib.pyplot as plt
import numpy as np

n = np.arange(0, 2500, 1)

plt.plot(n, A(n))

plt.title('Relación entre el tiempo de ejecución de un algoritmo y la cantidad de elementos que ordena')
plt.xlabel('Cantidad de elementos (unidades)')
plt.ylabel('Tiempo de ejecución (segundos)')

plt.grid(True)
plt.show()
```

$6$ horas de funconamiento equivalen a $6\cdot 3.600 = 21.600$ segundos.

Reemplazando en la función se tiene:

$21.600 = 0,01n^{2} + 0,5n + 2$

$0 = 0,01n^{2} + 0,5n + 2 - 21.600$

$0 = 0,01n^{2} + 0,5n - 21.598$

Ahora definimos la nueva función para calcular las soluciones

```
import numpy as np
from scipy.optimize import fsolve

#Definiendo la nueva función
def tiempo(n):
  return 0.01 * n**2 + 0.5 * n - 21598

n = np.linspace(0, 5000, 10)

solucion = fsolve(tiempo, n)

print(f'A las 6 horas de funcionamiento el algoritmo ha ordenado {solucion[0]:.0f} elementos aproximadamente.')
```

# Problema 9

El consumo de energía (en $kWh$) de un *Data Center* puede ser modelado por la función:

 $$E(t) = 50\cdot log(t + 1) + 200$$

donde $t$ corresponde al tiempo (en horas) desde el inicio del monitoreo.

1. Defina variable dependiente e independiente, indicando unidad de medida.
2. Determine el consumo de energía del *Data Center* después de $5$ horas.
3. ¿Luego de cuántas horas el consumo será de $350$ kWh?
4. Grafique, utilizando la biblioteca *Matplotlib*, la función $E(t)$.

## Solución

**1)**

**Variable dependiente**: Energía consumida (kWk)

**Variable independiente**: Tiempo transcurrido (horas)

**2)**


```
def E(t):
  return 50 * np.log10(t + 1) + 200

print(f'Transcurridas 5 horas el Data Center ha consumido {E(5):.0f} kWh aproximadamente.')
```

**3)**

Sabemos que $E(t) = 350$, reemplazamos este valor en la función e igualamos a $0$:

$350 = 50\cdot log(t + 1) + 200$

$0 = 50\cdot log(t + 1) + 200 - 350$

$0 = 50\cdot log(t + 1) - 150$

Ahora definimos esta nueva función para calcular con *fsolve*.



```
import numpy as np
from scipy.optimize import fsolve

#Definimos la nueva función
def energia(t):
  return 50 * np.log10(t + 1) - 150

t = np.linspace(0, 30, 1)
solucion = fsolve(energia, t)

print(f'Si el Data Center ha consumido 350 kWh es porque ha funcionado {solucion[0]:.0f} horas aproximadamente.')
```

**4)** Gráfico de la función

```
import matplotlib.pyplot as plt
import numpy as np

def E(t):
  return 50 * np.log10(t + 1) + 200

t = np.arange(0, 1000, 1)

plt.plot(t,E(t))

plt.title('Relación entre la energía consumida por el Data Center y el tiempo transcurrido')
plt.xlabel('Tiempo transcurrido (horas)')
plt.ylabel('Energía consumida (kWh)')

plt.grid(True)
plt.show()
```

# Problema 10

En gestión de proyectos, es crucial entender cómo se distribuye la carga de trabajo a lo largo del tiempo para planificar recursos, tiempo y esfuerzos de manera eficiente. Un fenómeno común es la disminución exponencial de la carga de trabajo, donde el esfuerzo requerido es mayor al inicio del proyecto y disminuye gradualmente a medida que se completan las tareas principales.

Esta disminución exponencial puede modelarse matemáticamente para predecir y gestionar el trabajo de manera efectiva.

<br>

Un estudio sobre gestión de proyectos sostiene que la carga de trabajo en un proyecto (en porcentaje) puede ser modelada por la función:

$$W(t) = 100e^{-0.1t}$$

donde $t$ corresponde al tiempo transcurrido (en semanas) desde el inicio del proyecto.

1. Defina variable dependiente e independiente, indicando unidad de medida.
2. Determine la carga de trabajo al inicio del proyecto.
3. ¿Cuál será la carga de trabajo luego de transcurridas $4$ semanas desde el inicio del proyecto?
4. Si el proyecto duró $12$ semanas, ¿es correcto afirmar que la carga de trabajo llegó al $20\%$?
5. ¿Cuántas semanas han pasado desde el inicio del proyecto para que la carga de trabajo sea de $55\%$?
6. Utilizando *Matplotlib*, grafique la función considerando un tiempo máximo de $12$ semanas.

## Solución

**1)**

**Variable dependiente**: Carga de trabajo (porcentaje)

**Variable independiente**: Tiempo transcurrido (semanas)

**2)**

Al inicio del proyecto $t = 0$.

```
def W(t):
  return 100 * np.exp(-0.1 * t)

print(f'La carga de trabajo al inicio del proyecto es {W(0):.0f}%.')
```

**3)**


```
print(f'Transcurridas 4 semanas la carga de trabajo es {W(4):.0f}%.')
```

**4)**

```
print(f'{W(12)}\n')

print(f'Transcurridas 12 semanas desde el inicio del proyecto la carga de trabajo es {W(12):.0f}%, por lo tanto, no es correcto afirmar que la carga es del 20%.')
```

**5)**

$W(t) = 55$

$55 = 100e^{-0,1t}$

$0 = 100e^{-0,1t} - 55$

```
import numpy as np
from scipy.optimize import fsolve

#Definimos la nueva función
def w(t):
  return 100 * np.exp(-0.1 * t) - 55

t = np.linspace(0, 20, 1)
solucion = fsolve(w, t)

print(f'Si la carga de trabajo es de 55% es porque han transcurrido {solucion[0]:.0f} semanas aproximadamente desde el inicio del proyecto.')

```

**6)** Gráfico de la función

```
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0, 13, 0.1)

plt.plot(t, W(t))

plt.title('Relación entre la carga de trabajo y el tiempo transcurrido')
plt.xlabel('Tiempo transcurrido (semanas)')
plt.ylabel('Carga de trabajo (porcentaje)')

plt.grid(True)
plt.show()
```

Un equipo de TI está monitoreando el tráfico de datos en la red de la empresa. Han registrado el volumen de los datos transferidos y el tiempo que tomó cada transferencia en la siguiente tabla de datos:

</center></p>

2. Utilizando la función *polyfit* de *Numpy* encuentra la función que modela el tiempo de transferencia $T(x)$ en función del volumen de los datos transferidos.
3. Interprete la pendiente en el contexto del problema.
4. ¿Cuánto tiempo toma una transferencia de $73,2$ GB?
5. Utiliza la función *fsolve* de *Scipy* para determinar la cantidad de datos que se puedan transferir en $123,5$ minutos.

## Solución

**1)**


**Variable dependiente:** Tiempo de transferencia (minutos)

**Variable independiente:** Volumen de los datos (GB)


**2)**

```
import numpy as np

datos = np.array([5, 10, 25, 50, 100])
tiempo = np.array([10, 20, 50, 100, 200])

pendiente, intercepto = np.polyfit(datos, tiempo, 1)
print(f'La pendiente de la función es {pendiente} y el intercepto es {intercepto}.')
```

La función que modela el problema es:

$$T(x) = 2x$$

**3)**

La pendiente indica que por cada GB transferido el tiempo de la transferencia aumenta en $2$ minutos.

**4)**

```
def T(x):
  return 2 * x

print(f'Si el volumen de la transferencia es de 73,2 GB el tiempo de la transferencia es {T(73.2)} minutos.')
```

**5)**

Cuando el tiempo de transferencia es de $123,5$ minutos:

$T(x) = 123,5$

$123,5 = 2x$

$0 = 2x - 123,5$

## Solución

**1)**

La función $I(x)$ indica que al inicio de la campaña el ingreso por ventas es de $\$1.000.000$. A medida que las semanas aumentan, el ingreso por ventas disminuye. Específicamente, por cada semana de campaña, el ingreso por ventas disminuye en $20$ mil pesos.

<br>

Donde $x$ corresponde al número de suscriptores (en miles).

## Solución

**1)**

El punto de equilibrio ocurre cuando:

\begin{split}
I(x) & = C(x)\\
0 & = I(x) - C(x)
\end{split}


```
import numpy as np
from scipy.optimize import fsolve

#Ingreso por suscriptores
def I(x):
  return 10 * x**2 + 50 * x

#Costos de operación
def C(x):
  return 5 * x**2 + 80 * x + 100

#Definiendo la función igualada a cero
def interseccion(x):
  return I(x) - C(x)

x = np.linspace(0, 10, 2)
solucion = np.around(fsolve(interseccion, x), decimals = 1)
suscriptores = np.unique(solucion)

suscriptor = [elementos for elementos in suscriptores if elementos >= 0]

print(f'El punto de equilibrio ocurre en x = {suscriptores}.')
print(f'Cuando la plataforma tiene {suscriptor[0]} mil suscriptores, el ingreso por suscriptores es igual a los costos de operación: {I(suscriptor[0]):.2f} mil pesos.')
```

0 & = S_{oferta}(x) - S_{actual}(x)
\end{split}


```
import numpy as np
from scipy.optimize import fsolve

#Definiendo la función igualada a cero
def interseccion(x):
  return S_oferta(x) - S_actual(x)

x = np.linspace(0, 40, 2)
solucion = np.around(fsolve(interseccion, x), decimals = 0)
casas = np.unique(solucion)

sol_casas = [elementos for elementos in casas if elementos >= 0]

print(f'El punto de equilibrio ocurre en x = {casas}.')
print(f'Cuando Constanza vende {sol_casas[0]:.0f} casas, el sueldo es igual en ambas empresas: {S_actual(sol_casas[0]):.1f} mil pesos.')
```

Le aconsejaría que si en promedio vende menos de $10$ casas mensuales, debe quedarse en su actual trabajo. Si vende más de $10$ casas en promedio al mes, es mejor que acepte la oferta.

**1)** Gráfico de las funciones

```
import matplotlib.pyplot as plt
import numpy as np

#Primera especie de algas
def f(x):
  return np.sqrt(500 * x + 2000)

#Segunda especie de algas
def g(x):
  return np.sqrt(800 * x + 1000)

x = np.arange(0, 500, 1)

plt.plot(x, f(x), color = 'm', label = 'Primera especie: f(x)')
plt.plot(x, g(x), color = 'y', label = 'Segunda especie: g(x)')
plt.ylim(0, 150)
plt.xlim(0, 20)

plt.title('Relación entre la tasa de crecimiento de dos especies de algas y la intensidad de luz recibida')
plt.xlabel('Intensidad de luz (lux)')
plt.ylabel('Tasa de crecimiento (mm/día)')

plt.grid(True)
plt.legend()
plt.show()
```

\begin{split}
f(x) & = g(x)\\
0 & = g(x) - f(x)
\end{split}


```
import numpy as np
from scipy.optimize import fsolve

#Definiendo la función igualada a cero
def interseccion(x):
  return g(x) - f(x)

x = np.linspace(0, 10, 2)
solucion = np.around(fsolve(interseccion, x), decimals = 2)
intensidad = np.unique(solucion)

print(f'El punto de intersección ocurre en x = {intensidad}.')
print(f'Cuando la intensidad de la luz es {intensidad[0]} lux, la tasa de crecimiento es igual en ambas especies: {f(intensidad[0]):.2f} mm/día.')
```

# Problema 10

En una planta de tratamiento de aguas residuales, los ingenieros están estudiando la eficiencia, medida en porcentaje, de dos diferentes procesos de filtración para eliminar contaminantes. La eficiencia de cada proceso varía en función del tiempo de tratamiento, $t$, medido en horas.

Las funciones que describen la eficiencia de los procesos son las siguientes:

Proceso A:

 $$A(t) = 1,4t^{4} - 2,55t^{3} - 3,37t^{2} + 2,75t + 10$$

Proceso B:

 $$B(t) = t^{4} -2,67t^{3} - t^{2} + 2t + 20$$

 <br>

1. Grafique ambas funciones indicando el nombre de cada eje coordenado, junto con su unidad de medida. Utilice *Matplotlib*.
2. Transcurridas $3$ horas y media, ¿qué porcentaje de eficiencia tiene el proceso A? ¿Y el B?
3. ¿Para cuántas horas transcurridas el proceso A tiene la misma eficiencia que el proceso B? Indique la eficiencia en este punto.
4. Si sabe que los procesos demoran como máximo $5$ horas, analice cuándo es conveniente aplicar el proceso A.

## Solución

**1)** Gráfico de las funciones

```
import matplotlib.pyplot as plt
import numpy as np

#Proceso A
def A(t):
  return 1.4 * t**4 - 2.55 * t**3 - 3.37 * t**2 + 2.75 * t + 10

#Proceso B
def B(t):
  return t**4 - 2.67 * t**3 - t**2 + 2 * t + 20

t = np.arange(-100, 5000, 0.1)

plt.plot(t, A(t), color = 'lime', label = 'Proceso A')
plt.plot(t, B(t), color = 'deeppink', label = 'Proceso B')
plt.ylim(0, 150)
plt.xlim(-5,5)

plt.title('Relación entre la eficiencia de dos procesos y el tiempo transcurrido')
plt.xlabel('Tiempo transcurrido (horas)')
plt.ylabel('Eficiencia del proceso (porcentaje)')

plt.grid(True)
plt.legend()
plt.show()
```

**2)**


```
print(f'Transcurridas 3,5 horas, el proceso A tiene una eficiencia del {A(3.5):.1f}% y el B de {B(3.5):.1f}%.')
```

**3)**

La intersección ocurre cuando:

\begin{split}
A(t) & = B(t)\\
0 & = B(t) - A(t)
\end{split}




```
import numpy as np
from scipy.optimize import fsolve

#Definiendo la función igualada a cero
def interseccion(t):
  return B(t) - A(t)

t = np.linspace(-10, 10, 4)
solucion = np.around(fsolve(interseccion, t), decimals = 1)
unica = np.unique(solucion)

tiempo = [elementos for elementos in unica if elementos >= 0]

print(f'El punto de intersección ocurre en t = {tiempo}.')
print(f'Cuando han transcurrido {tiempo[0]} horas, ambos procesos muestran la misma eficiencia: {A(tiempo[0]):.1f}%.  ')
```

**4)**

El proceso A es más eficiente si se sabe que el tratamiento debe durar más de $2,8$ horas. Para tratamientos que duren menos de $2,8$ horas es conveniente usar el proceso B.

```
import matplotlib.pyplot as plt
import numpy as np

def tramo_1(x):
  return 2 - x

def tramo_2(x):
  return 3

#Vectorizamos el tramo 2 para que tenga las mismas dimensiones que "x"
tramo_2 = np.vectorize(tramo_2)

x = np.arange(-50, 50, 0.001)
f_x = np.piecewise(x, [x < 1, x >= 1], [lambda x: tramo_1(x), lambda x: tramo_2(x)])


plt.plot(x[x < 1], tramo_1(x[x < 1]), color = 'darkred', label = 'Tramo 1')
plt.plot(x[x >= 1], tramo_2(x[x >= 1]), color = 'c', label = 'Tramo 2')

plt.title('Gráfico de f(x)')

plt.grid(True)
plt.legend()
plt.show()
```
